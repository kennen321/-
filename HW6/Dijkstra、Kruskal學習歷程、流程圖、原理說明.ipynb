## Dijkstra原理說明

Dijkstra(戴克斯特拉算法)是用來算"Shortest Path"

使用了廣度優先搜尋解決賦權有向圖的單源最短路徑問題。

該演算法存在很多變體；戴克斯特拉的原始版本找到兩個頂點之間的最短路徑，

但是更常見的變體固定了一個頂點作為源節點然後找到該頂點到圖中所有其它節點的最短路徑，產生一個最短路徑樹。

該演算法常用於路由演算法或者作為其他圖演算法的一個子模組。

舉例來說，如果圖中的頂點表示城市，而邊上的權重表示城市間開車行經的距離，該演算法可以用來找到兩個城市之間的最短路徑。

用最原始的方式實現dijkstra，就是每次從costs裡面找最短路徑的點，

再遍歷這個點的邊，更新最短路徑。由於每次都要從costs裡面找最短路徑，時間複雜讀為O(n^2)。

## Kruskal原理說明

```python
Kruskal(克魯斯克爾演算法)
一個有 n 個結點的連通圖的生成樹是原圖的極小連通子圖，且包含原圖中的所有 n 個結點，並且有保持圖連通的最少的邊。

最小生成樹可以用kruskal（克魯斯卡爾）算法或prim（普裏姆）算法求出。最小生成樹其實是最小權重生成樹的簡稱。

是一種用來尋找最小生成樹的算法，在剩下的所有未選取的邊中，找最小邊，如果和已選取的邊構成回路，則放棄，選取次小邊。
```

```python
簡單來說，Kruskal是以增加邊的觀念做為出發點。

先將所有的邊，依照權重的大小排序。再來依序加入權重最小的邊，

如果 造成cycle時，則必須捨棄，直到增加了n - 1條邊為止。(假設有 n 個節點)
```
